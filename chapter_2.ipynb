**1-4 if Statements**
The most common type of flow control statement is the if statement. An if statements clause (that is, the block following the if statement) will execute if the statements condition is True. The clause is skipped if the condition is False.

**1-4-1 elif Statements** :While only one of the if or else clauses will execute, you may have a case where you want one of many possible clauses to execute. The elif statement is an else if statement that always follows an if or another elif statement

```python
nawe_2 = "lolo"
age = 24
if nawe_2 == "lolo":
  print("hello lolo")
elif age <  10 :
  print("is not your age ")
elif age > 10:
  print("this is your age ")
```
