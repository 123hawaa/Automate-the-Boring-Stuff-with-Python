# Chapter 10: Reading and Writing Files



## Section 1: File Paths
File paths tell Python where a file or folder is located. We can use the `os` module to work with them in a cross-platform way.

Operating systems organize files into folders (also called directories).

**Modules**: os and Path modules

**Path**: Is used to describe location of the file (absolute: is_absolute() & relative: is_relative. Returns boolean values; True or False)

**Filename**: Name of the the file

```python
import os

print("Current Working Directory:", os.getcwd())
# Create a cross-platform path using os.path.join()
path = os.path.join("folder", "subfolder", "file.txt")
print("Example Path:", path)
```

## Section 2: Reading Files
We can open files in **read mode** (`'r'`) and use `.read()` to read the entire file or iterate line by line.

```python
# Create a sample text file to read
with open("example.txt", "w") as f:
    f.write("Hello, world!\nThis is a test file.\nEnjoy learning Python!")

# Read entire file
file = open("example.txt", "r")
print("File content using read():")
print(file.read())
file.close()

# Read line by line
with open("example.txt", "r") as f:
    print("\nFile content line by line:")
    for line in f:
        print(line.strip())
```

```python
import os

# Example paths
absolute_path = "/content/my_folder/my_file.txt"
relative_path = "my_folder/my_file.txt"
drive_path = "/gdrive/My Drive/my_colab_project/data.csv" # Absolute path after mounting Google Drive

# Check if paths are absolute
print(f"Is '{absolute_path}' absolute? {os.path.isabs(absolute_path)}")
print(f"Is '{relative_path}' absolute? {os.path.isabs(relative_path)}")
print(f"Is '{drive_path}' absolute? {os.path.isabs(drive_path)}")
```

## Section 3: Writing Files
We can open files in **write mode** (`'w'`) to overwrite content or **append mode** (`'a'`) to add new content.

```python
# Overwrite the file
with open("example.txt", "w") as f:
    f.write("This text overwrote the old content.\n")

# Append new content
with open("example.txt", "a") as f:
    f.write("This line was appended.\n")

# Show final content
with open("example.txt", "r") as f:
    print("\nAfter writing and appending:")
    print(f.read())
```

## Section 4: Using `with` Statement
The `with` statement is safer because it automatically closes the file after use, even if an error occurs.

```python
with open("with_example.txt", "w") as f:
    f.write("Using the 'with' statement is safer!")

with open("with_example.txt", "r") as f:
    print("With statement example:")
    print(f.read())
```

## Section 5: Shelve Module
The `shelve` module stores Python objects (like lists or dictionaries) into a file, working like a persistent dictionary.

```python
import shelve

with shelve.open("mydata") as shelf:
    shelf["cats"] = ["Zophie", "Pooka", "Simon"]

with shelve.open("mydata") as shelf:
    print("Stored with shelve:", shelf["cats"])
```

## Section 6: Pprint Module
The `pprint` module makes data structures easier to read and can be used to save them in a human-readable format.

```python
import pprint
cats = [{"name": "Zophie", "age": 7}, {"name": "Pooka", "age": 5}]

with open("cats.txt", "w") as f:
    f.write(pprint.pformat(cats))

print("Data saved with pprint:")
with open("cats.txt", "r") as f:
    print(f.read())
```
